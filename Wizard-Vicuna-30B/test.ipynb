```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
```

```python
tokenizer = AutoTokenizer.from_pretrained("./Wizard-Vicuna-30B-Uncensored-GPTQ", device_map='cuda')
model = AutoModelForCausalLM.from_pretrained("./Wizard-Vicuna-30B-Uncensored-GPTQ", device_map='cuda')
device = "cuda:0" if torch.cuda.is_available() else "cpu"
model = model.to(device)
```

```python
system = "You are a psychology tutor"
user = "White an essay about the different types of neurons with at least 1000 words"
input_text = "<|start_header_id|>system<|end_header_id|>" + system + "<|eot_id|><|start_header_id|>user<|end_header_id|>" + user + "<|eot_id|><|start_header_id|>assistant<|end_header_id|>"
```

```python
inputs = tokenizer(input_text, return_tensors="pt").to(device)
outputs = model.generate(**inputs, max_length = 2048)
generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(generated_text)
```
